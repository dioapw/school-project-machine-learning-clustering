# **K-Means Clustering**

Pertama kita hanya akan menggunakan dua fitur 'Umur', 'Premi' sehingga kita dapat memvisualisasikan cluster pada bidang 2D

```python
data = df.loc[:, ['Umur', 'Premi']]
data.head(2)
```

Konversikan ke array numpy

```python
X = data.values
```

Visualisasikan titik data

```python
sns.scatterplot(x=X[:,0], y=X[:, 1])

plt.xlabel('Umur')
plt.ylabel('Premi')
plt.show()
```

Kalkulasi WCSS

```python
def calculate_cost(X, centroids, cluster):  
    sum = 0  
    for i, val in enumerate(X):    
        sum += np.sqrt((centroids[int(cluster[i]), 0] - val[0])**2 +(centroids[int(cluster[i]), 1] - val[1])**2)  
    return sum
```

Menerapkan K-Means

```python
def kmeans(X, k):  

    diff = 1  
    cluster = np.zeros(X.shape[0])  
    centroids = data.sample(n=k).values  

    while diff:     

        # for each observation     
        for i, row in enumerate(X):  

            mn_dist = float('inf')   

            # dist of the point from all centroids        
            for idx, centroid in enumerate(centroids):    

                    d = np.sqrt((centroid[0]-row[0])**2 + (centroid[1]-row[1])**2)    

                    # store closest centroid            
                    if mn_dist > d:               
                        mn_dist = d               
                        cluster[i] = idx     
                    
        new_centroids = pd.DataFrame(X).groupby(by=cluster).mean().values

        # if centroids are same then leave     
        if np.count_nonzero(centroids-new_centroids) == 0:        
            diff = 0      
        else:        
            centroids = new_centroids  
            
    return centroids, cluster
```

Cari nilai K menggunakan metode elbow

```python
cost_list = []

for k in range(1, 8): 

    centroids, cluster = kmeans(X, k)   
     
    # WCSS (Within cluster sum of square)    
    cost = calculate_cost(X, centroids, cluster)    
    cost_list.append(cost)
```

Buat plot garis antara WCSS dan k

```python
sns.lineplot(x=range(1,8), y=cost_list, marker='o')

plt.xlabel('k')
plt.ylabel('WCSS')
plt.show()
```

WCSS tidak banyak berkurang setelah k=5, jadi buatlah 5 cluster

Membuat clusters

```python
k = 5
centroids, cluster = kmeans(X, k)
```

Visualisasikan cluster yang terbentuk

```python
sns.scatterplot(x=X[:,0], y=X[:, 1], hue=cluster)
sns.scatterplot(x=centroids[:,0], y=centroids[:, 1], s=100, color='y')
plt.xlabel('Umur')
plt.ylabel('Premi')
plt.show()
```
